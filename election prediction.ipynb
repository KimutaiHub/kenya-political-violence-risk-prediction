# Predicting the Severity of Political Violence in Kenya
### A Machine Learning Approach Using Historical Conflict Fatality Data

# 1. Overview

Predicting the Severity of Political Violence in Kenya is an end-to-end machine learning project that uses 29 years of ACLED conflict data (1997–2026), enriched with Kenya census population and density figures, to model fatality counts at the county-week level across all 47 counties. The dataset spans 16,549 observations and supports a full pipeline from data merging through modeling, interpretability, and a county-level vulnerability index.

## 1.1 Business Understanding

Kenya has experienced recurring episodes of political violence since the early 1990s, with the most severe flashpoints clustered around contested general elections — most notably the 2007/08 post-election crisis and the 2017 repeat presidential election. Political violence in Kenya is frequently linked to ethnic mobilization, contested results, and localized historical grievances that resurface predictably around electoral cycles.

Despite this recurring pattern, conflict response in Kenya has historically been **reactive rather than predictive**. Security deployments, mediation efforts, and humanitarian pre-positioning tend to follow violence rather than anticipate it. This is not due to a lack of data — organizations like ACLED (Armed Conflict Location & Event Data Project) have systematically tracked conflict events, locations, actors, and fatalities across Kenya for nearly three decades. The gap is in **translating that raw event data into structured, actionable, forward-looking risk intelligence** at a level decision-makers can use: the county.

### Why This Matters Now

Kenya's next general election cycles will continue to follow the same five-year rhythm that has historically produced spikes in political violence. Stakeholders such as the National Cohesion and Integration Commission (NCIC), County Security Committees, election observer bodies (e.g. ELOG), and humanitarian organizations (e.g. UNHCR) all share a common need: knowing **where** violence is likely to concentrate and **how severe** it might become, early enough to act.


By combining nearly 30 years of granular ACLED conflict records with structural demographic data (population and density by county from the Kenya Census), this project applies machine learning to move from descriptive conflict reporting toward **predictive, county-level risk intelligence**. Rather than simply counting past violent events, the goal is to model the conditions under which violence becomes more severe — and to surface that insight in a form usable by people who are not data scientists: a ranked vulnerability index and interpretable feature drivers.

This positions the project at the intersection of **conflict studies, political science, and applied machine learning**, with a clear humanitarian and policy use case rather than a purely academic one.

## 1.2 Business Problem


Political violence in Kenya is not random — it is patterned by geography, event type, population density, and temporal cycles aligned with elections and resource conflicts. Despite this predictability, early warning systems and humanitarian resource allocation in Kenya continue to rely primarily on expert judgment and lagged reporting, rather than systematic, data-driven forecasting. By the time a response is mobilized, violence has often already escalated.

This creates a clear gap: rich historical conflict data exists (ACLED has tracked Kenyan political violence since 1997), but it has not been translated into a forward-looking, county-level decision-support tool.

**Formally, the business problem is:**

> Given historical patterns of conflict events, fatalities, and county-level demographic structure, can we predict the *severity* of political violence (fatalities) at the county-week level — well enough to flag high-risk counties before violence escalates, and to explain *why* those counties are at risk in terms decision-makers can act on?

A solution to this problem must address three sub-needs simultaneously:

1. **Quantification** — produce a numeric estimate of how many fatalities a given county-week is likely to see, based on conflict history, event patterns, and demographic conditions.
2. **Early warning** — surface high-risk counties *before* violence escalates, not after, so resources can be pre-positioned rather than reactively deployed.
3. **Interpretability** — go beyond a "black box" prediction and provide feature-level explanations (e.g. which factors — population density, prior unrest, election proximity — are driving the risk) so policy actors can justify and target interventions.

**Why this is a hard problem:** the target variable, `FATALITIES` per county-week, is heavily right-skewed — 72.6% of observations are zero, while the maximum recorded value is 281 fatalities in a single county-week. This means the events that matter most (severe, high-fatality outbreaks) are also the rarest in the data, making this fundamentally a problem of **predicting rare, high-impact outcomes from a sparse and noisy signal** — both analytically non-trivial and practically urgent, since these rare events are precisely the ones early-warning systems exist to catch.
